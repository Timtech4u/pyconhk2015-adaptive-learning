## Logistic Function

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def logistic(x, a):
    return 1 / (1 + np.exp(-a * (x - 5)))

plt.figure(figsize=(14,4))
x = np.linspace(0, 10, 1000)

plt.subplot(1, 2, 1)
plt.plot(x, logistic(x, 1))
plt.xlim((0,10))
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('a = 1')

plt.subplot(1, 2, 2)
plt.plot(x, logistic(x, 2))
plt.xlim((0,10))
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('a = 2')
plt.show()
```

## Building the Bayesian Model

```python
import pymc as pm
import random
import math

# Some artificial data
# The learner will answer correctly 70% of the time
data = []
for i in xrange(500):
    x = random.random()
    if (x > 0.3):
        data.append(1)
    else:
        data.append(0)

a = 1
b = 1
        
# Define theta with a uniform distribution
theta = pm.Beta('theta', a, b)

# p is obtained from a logistic function of theta and a
@pm.deterministic
def p(theta=theta):
    return 1.0 / (1 + math.exp(-1 * (theta * 10 - 5)))

# x is observed data, it is drawn from a Bernoulli distribution,
# with parameter p, i.e. P(x=1) = p
# 'data' contains observed data defined above, and we set observed=True
x = pm.Bernoulli('x', p, value=data, observed=True)
#x = pm.Binomial('x', n=len(data), p=p, value=data.count(1), observed=True)

# Define a pymc model with all the components
model = pm.Model([theta, p, x])
```

## Obtaining the MAP (Maximum a Posterior Probability)

```python
# Create a MAP (maximum a posteriori estimate) object
m = pm.MAP(model)
m.fit()

# Check the MAP estimate of the variables
print "Theta: %.4f" % m.get_node('theta').value
print "p    : %.4f" % m.get_node('p').value
```

## Using MCMC to Obtain Samples from Posterior Distributions

```python
# from the posterior probability distributions
mcmc = pm.MCMC(model)
mcmc.sample(iter=5000, burn=1000)
```
